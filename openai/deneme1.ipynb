```python
with open('ex5.txt', 'r', encoding='utf-8') as file:
    metin_verisi = file.read()
    print(metin_verisi)
```

```python
from nltk.tokenize import sent_tokenize

cumleler = sent_tokenize(metin_verisi)
```

```python
cumleler
```

```python
from nltk.tokenize import word_tokenize

kelimeler = [word_tokenize(cumle) for cumle in cumleler]
```

```python
kelimeler
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('turkish'))
```

```python
stop_words
```

```python
temiz_kelimeler = []

for cumle in cumleler:
    kelimeler = word_tokenize(cumle)
    for kelime in kelimeler:
        if kelime.lower() not in stop_words:
            temiz_kelimeler.append(kelime.lower())
```

```python
temiz_kelimeler
```

```python
from collections import Counter

kelime_sayilari = Counter(temiz_kelimeler)
en_sik_kelimeler = kelime_sayilari.most_common(10)

print("En sık kullanılan 10 kelime:")
for kelime, sayi in en_sik_kelimeler:
    print(kelime, ":", sayi)
```

```python
import string

temiz_kelimeler = []

for cumle in cumleler:
    kelimeler = word_tokenize(cumle)
    for kelime in kelimeler:
        kelime = kelime.lower()
        if kelime not in stop_words and kelime not in string.punctuation:
            temiz_kelimeler.append(kelime)
```

```python
from collections import Counter

kelime_sayilari = Counter(temiz_kelimeler)
en_sik_kelimeler = kelime_sayilari.most_common(10)

print("En sık kullanılan 10 kelime:")
for kelime, sayi in en_sik_kelimeler:
    print(kelime, ":", sayi)
```
