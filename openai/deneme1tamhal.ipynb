```python
with open('ex5.txt', 'r', encoding='utf-8') as file:
    metin_verisi = file.read()
    print(metin_verisi)
from nltk.tokenize import sent_tokenize

cumleler = sent_tokenize(metin_verisi)
from nltk.tokenize import word_tokenize

kelimeler = [word_tokenize(cumle) for cumle in cumleler]
from nltk.corpus import stopwords

stop_words = set(stopwords.words('turkish'))
temiz_kelimeler = []

for cumle in cumleler:
    kelimeler = word_tokenize(cumle)
    for kelime in kelimeler:
        if kelime.lower() not in stop_words:
            temiz_kelimeler.append(kelime.lower())
```
